# Step 1: Importing data 

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt   
import seaborn as sns     
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer= load_breast_cancer()
```

```python
cancer.keys()     
```

```python
print(cancer['target'])
```

```python
print(cancer['feature_names'])
```

```python
print(cancer['data'].shape)   
```

```python
df_cancer=pd.DataFrame(np.c_[cancer['data'],cancer['target']], columns=np.append(cancer['feature_names'],['target']))

'''first it will take all the 30 columns of the dataset by data and then include one additional column named target which is the
31st column and then we append 2 vectors into columnn'''

df_cancer.head()
```

# Step2: Visualizing the data

```python
sns.pairplot(df_cancer, hue='target',vars=['mean radius','mean texture','mean perimeter','mean area','mean smoothness'])

```

```python
sns.countplot(df_cancer['target'])   
```

```python
sns.scatterplot(x='mean area', y ='mean smoothness', hue='target',data=df_cancer)  
```

```python
plt.figure(figsize=(20,10)) 
sns.heatmap(df_cancer.corr(),annot= True) 
```

# Step 4: Model training (finding a prolem solution)


```python
X = df_cancer.drop(['target'], axis=1)
X
```

```python
y=df_cancer['target']
y
```

```python
#splitting data into training and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(X_train,y_train)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

# Step 5: Evaluating the model

```python
y_predict=svc.predict(X_test)
y_predict
```

```python
cm= confusion_matrix(y_test,y_predict)
```

```python
sns.heatmap(cm,annot=True)
```

# Step 6: Improving the model

```python
#Feature Scaling
```

```python
min_train=X_train.min()
max_train=X_train.max()
#range_train=(X_train-min_train).max()               #data normalization===== X'=X-X(min)/X(max)-X(min)
#X_train_scaled=(X_train-min_train)/range_train
X_train_scaled=(X_train-min_train)/(max_train-min_train)
```

```python
sns.scatterplot(x=X_train_scaled['mean area'], y =X_train_scaled['mean smoothness'], hue=y_train)  
```

```python
min_test=X_test.min()
max_test=X_test.max()
X_test_scaled=(X_test-min_test)/(max_test-min_test)
```

```python
svc.fit(X_train_scaled,y_train)
```

```python
y_predict=svc.predict(X_test_scaled)
y_predict
```

```python
cm= confusion_matrix(y_test,y_predict)
cm
```

```python
sns.heatmap(cm,annot=True)
```

```python
print(classification_report(y_test,y_predict))
```

#  Step 7: Improving the model (Part - 2)

```python
param_grid={'C':[0.1,1,10,100] , 'gamma':[1,0.1,0.01,0.001],'kernel':['rbf']}     
```

```python
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(SVC(),param_grid,refit=True,verbose=4)
grid.fit(X_train_scaled,y_train)
```

```python
#to get our best value
grid.best_params_
```

```python
grid_predictions=grid.predict(X_test_scaled)
```

```python
cm= confusion_matrix(y_test,grid_predictions)
cm
```

```python
sns.heatmap(cm,annot=True)
```

```python
print(classification_report(y_test,grid_predictions))
```
